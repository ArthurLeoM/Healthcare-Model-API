```python
class Sparsemax(nn.Module):
    """Sparsemax function."""

    def __init__(self, dim=None):
        """Initialize sparsemax activation
        
        Args:
            dim (int, optional): The dimension over which to apply the sparsemax function.
        """
        super(Sparsemax, self).__init__()

        self.dim = -1 if dim is None else dim

    def forward(self, input):
        """Forward function.

        Args:
            input (torch.Tensor): Input tensor. First dimension should be the batch size

        Returns:
            torch.Tensor: [batch_size x number_of_logits] Output tensor

        """
        # Sparsemax currently only handles 2-dim tensors,
        # so we reshape and reshape back after sparsemax
        original_size = input.size()
        input = input.view(-1, input.size(self.dim))
        
        dim = 1
        number_of_logits = input.size(dim)

        # Translate input by max for numerical stability
        input = input - torch.max(input, dim=dim, keepdim=True)[0].expand_as(input)

        # Sort input in descending order.
        # (NOTE: Can be replaced with linear time selection method described here:
        # http://stanford.edu/~jduchi/projects/DuchiShSiCh08.html)
        zs = torch.sort(input=input, dim=dim, descending=True)[0]
        range = torch.arange(start=1, end=number_of_logits+1, device=device, dtype=torch.float32).view(1, -1)
        range = range.expand_as(zs)

        # Determine sparsity of projection
        bound = 1 + range * zs
        cumulative_sum_zs = torch.cumsum(zs, dim)
        is_gt = torch.gt(bound, cumulative_sum_zs).type(input.type())
        k = torch.max(is_gt * range, dim, keepdim=True)[0]

        # Compute threshold function
        zs_sparse = is_gt * zs

        # Compute taus
        taus = (torch.sum(zs_sparse, dim, keepdim=True) - 1) / k
        taus = taus.expand_as(input)

        # Sparsemax
        self.output = torch.max(torch.zeros_like(input), input - taus)

        output = self.output.view(original_size)

        return output

    def backward(self, grad_output):
        """Backward function."""
        dim = 1

        nonzeros = torch.ne(self.output, 0)
        sum = torch.sum(grad_output * nonzeros, dim=dim) / torch.sum(nonzeros, dim=dim)
        self.grad_input = nonzeros * (grad_output - sum.expand_as(grad_output))

        return self.grad_input
```

```python
class CausalConv1d(torch.nn.Conv1d):
    def __init__(self,
                 in_channels,
                 out_channels,
                 kernel_size,
                 stride=1,
                 dilation=1,
                 groups=1,
                 bias=True):
        self.__padding = (kernel_size - 1) * dilation

        super(CausalConv1d, self).__init__(
            in_channels,
            out_channels,
            kernel_size=kernel_size,
            stride=stride,
            padding=self.__padding,
            dilation=dilation,
            groups=groups,
            bias=bias)

    def forward(self, input):
        result = super(CausalConv1d, self).forward(input)
        if self.__padding != 0:
            return result[:, :, :-self.__padding]
        return result
```

```python
class SELayer(nn.Module):
    def __init__(self, channel, reduction=9, use_h=True, use_c=True, activation='sigmoid'):
        super(SELayer, self).__init__()
        self.avg_pool = nn.AdaptiveAvgPool1d(1)
        self.use_h = use_h
        self.use_c = use_c
        scale_dim = 0
        self.activation = activation
        if use_h == True:
            self.nn_h = nn.Linear(channel, channel // reduction)
            scale_dim += channel // reduction
        if use_c == True:
            self.nn_c = nn.Linear(channel, channel // reduction)
            scale_dim += channel // reduction
        
        self.nn_rescale = nn.Linear(scale_dim, channel)
        self.sparsemax = Sparsemax(dim=1)

    def forward(self, x):
        b, c, t = x.size()

        y = None
        if self.use_c == True:
            y_origin = x[:, :, -1]
            se_c = self.nn_c(y_origin)
            se_c = torch.relu(se_c)
            y = se_c
        if self.use_h == True:
            y_pool = self.avg_pool(x).view(b, c) #B*C(*1)
            se_h = self.nn_h(y_pool)
            se_h = torch.relu(se_h)
            y = se_h if y is None else torch.cat((y, se_h), dim=-1)
        
        y = self.nn_rescale(y).view(b, c, 1)
        if self.activation == 'sigmoid':
            y = torch.sigmoid(y)
        else:
            y = self.sparsemax(y)
        return x * y.expand_as(x), y
```

```python
class MIDCare(nn.Module):
    def __init__(self, hidden_dim, kernel_size, kernel_num, input_dim, output_dim, dropout, r_v, r_c):
        super(MIDCare, self).__init__()
        self.hidden_dim = hidden_dim
        self.kernel_size = kernel_size
        self.kernel_num = kernel_num
        self.input_dim = input_dim
        self.output_dim = output_dim
        self.dropout = dropout
        
        self.nn_conv1 = CausalConv1d(input_dim, kernel_num, kernel_size, 1, 1)
        self.nn_conv3 = CausalConv1d(input_dim, kernel_num, kernel_size, 1, 4)
        self.nn_conv5 = CausalConv1d(input_dim, kernel_num, kernel_size, 1, 8)
        torch.nn.init.xavier_uniform_(self.nn_conv1.weight)
        torch.nn.init.xavier_uniform_(self.nn_conv3.weight)
        torch.nn.init.xavier_uniform_(self.nn_conv5.weight)
            
        self.nn_convse = SELayer(3*kernel_num, r_c, use_h=False, use_c=True, activation='sigmoid')  
        self.nn_inputse = SELayer(input_dim, r_v, use_h=False, use_c=True, activation='sparsemax')
        self.rnn = nn.GRUCell(input_dim+3*kernel_num, hidden_dim)
        self.nn_output = nn.Linear(hidden_dim, output_dim)
        self.nn_dropout = nn.Dropout(dropout)
        self.nn_dropoutc = nn.Dropout(0.3)
        
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()
        self.tanh = nn.Tanh()        

    def forward(self, input, demo_input=None, time_input=None):
        # input shape [batch_size, timestep, feature_dim]
        batch_size = input.size(0)
        time_step = input.size(1)
        feature_dim = input.size(2)

        cur_h = Variable(torch.zeros(batch_size, self.hidden_dim)).to(device)  
        inputse_att = []
        convse_att = []
        h = []
        
        conv_input = input.permute(0, 2, 1)
        conv_res1 = self.nn_conv1(conv_input)
        conv_res3 = self.nn_conv3(conv_input)
        conv_res5 = self.nn_conv5(conv_input)

        conv_res = torch.cat((conv_res1, conv_res3, conv_res5), dim=1) #B*3k*T
        conv_res = self.relu(conv_res)
        conv_res = self.nn_dropoutc(conv_res)
        
        for cur_time in range(time_step):
            convse_res, cur_convatt = self.nn_convse(conv_res[:, :, :cur_time+1])
            inputse_res, cur_inputatt = self.nn_inputse(input[:, :cur_time+1, :].permute(0, 2, 1))
            cur_input = torch.cat((convse_res[:, :, -1], inputse_res[:, :, -1]), dim=-1)
            
            cur_h = self.rnn(cur_input, cur_h)
            h.append(cur_h)
            convse_att.append(cur_convatt)
            inputse_att.append(cur_inputatt)
        
        h = torch.stack(h).permute(1,0,2)
        h_reshape = h.contiguous().view(batch_size * time_step, self.hidden_dim)
        if self.dropout > 0.0:
            h_reshape = self.nn_dropout(h_reshape)
        output = self.nn_output(h_reshape)
        output = self.sigmoid(output)
        output = output.contiguous().view(batch_size, time_step, self.output_dim)
        return output, convse_att, inputse_att
```

```python
class Fold_dataset(data.Dataset):
    def __init__(self, x, y, time, static, pdid, weight):
        self.x = x
        self.y = y
        self.time = time
        self.static = static
        self.pdid = pdid
        self.weight = weight

    def __getitem__(self, index):#返回的是tensor
        return self.x[index], self.y[index], self.time[index], self.static[index], self.weight[index], self.pdid[index]

    def __len__(self):
        return len(self.x)
```

```python
def get_loss(pred, true, mask):
    masked_output = pred * mask
    loss_task = true * torch.log(masked_output + 1e-7) + (1 - true) * torch.log(1 - masked_output + 1e-7)
    loss_task = torch.sum(loss_task, dim=1) / torch.sum(mask, dim=1)
    loss_task = torch.neg(torch.mean(loss_task))
    return loss_task
```

```python
epochs = 200
batch_size = 256

fold_count = 0
total_train_loss = []
total_valid_loss = []
global_best = 0
file_name = './model/midcare-random'
auroc = []
auprc = []
minpse = []

for train, test in kfold.split(x, y_mask):    
    model = MIDCare(hidden_dim, kernel_size, kernel_num, input_dim, output_dim, dropout, r_v, r_c).to(device)

    optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
    
    fold_count += 1
    max_len = max(len(x[i]) for i in train)
    x_train = np.zeros((len(train), max_len, 17), dtype=np.float32)
    y_train = np.zeros((len(train), max_len), dtype=np.float32)
    time_train = np.zeros((len(train), max_len), dtype=np.float32)
    weight_train = np.zeros((len(train), max_len), dtype=np.float32)
    static_train = np.array(static, dtype=np.float32)[train]
    pdid_train = np.array(pdid, dtype=np.float32)[train]
    
    #Padding
    for i in range(len(train)):
        x_train[i, :len(x[train[i]]), :] = x[train[i]]
        y_train[i, :len(y[train[i]])] = y[train[i]]
        time_train[i, :len(time[train[i]])] = time_new[train[i]]
        weight_train[i, :len(weight[train[i]])] = weight[train[i]]
    
    for i in range(len(x_train)):
        for j in range(len(x_train[i])):
            rdm = np.random.choice([0,1],p=np.array([0.2,0.8]).ravel())
            if rdm == 0 and j > 0:
                x_train[i, j, :] = x_train[i, j-1, :]
        
    train_dataset = Fold_dataset(x_train, y_train, time_train, static_train, pdid_train, weight_train)
    train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=False)
    
    max_len = max(len(x[i]) for i in test)
    x_test = np.zeros((len(test), max_len, 17), dtype=np.float32)
    y_test = np.zeros((len(test), max_len), dtype=np.float32)
    time_test = np.zeros((len(test), max_len), dtype=np.float32)
    weight_test = np.zeros((len(test), max_len), dtype=np.float32)
    static_test = np.array(static, dtype=np.float32)[test]
    pdid_test = np.array(pdid, dtype=np.float32)[test]

    #Padding
    for i in range(len(test)):
        x_test[i, :len(x[test[i]]), :] = x[test[i]]
        y_test[i, :len(y[test[i]])] = y[test[i]]
        time_test[i, :len(time[test[i]])] = time_new[test[i]]
        weight_test[i, :len(weight[test[i]])] = weight[test[i]]

    for i in range(len(x_test)):
        for j in range(len(x_test[i])):
            rdm = np.random.choice([0,1],p=np.array([0.2,0.8]).ravel())
            if rdm == 0 and j > 0:
                x_test[i, j, :] = x_test[i, j-1, :]
    
    test_dataset = Fold_dataset(x_test, y_test, time_test, static_test, pdid_test, weight_test)
    test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

    fold_train_loss = []
    fold_valid_loss = []
    best_auroc = 0
    best_auprc = 0
    best_minpse = 0
    
    for each_epoch in range(epochs):
        epoch_loss = []
        model.train()
        for step, (batch_x, batch_y, batch_time, batch_static, batch_weight, batch_pdid) in enumerate(train_loader):   
            optimizer.zero_grad()
            batch_x = batch_x.to(device)
            batch_y = batch_y.to(device)
            batch_time = batch_time.to(device)
            batch_static = batch_static.to(device)
            batch_weight = batch_weight.to(device)
                
            output,_,_ = model(batch_x, batch_time) #B T 1
            loss = get_loss(output, batch_y.unsqueeze(-1), batch_weight.unsqueeze(-1))
            epoch_loss.append(loss.cpu().detach().numpy())
            loss.backward()
            optimizer.step()
        epoch_loss = np.mean(epoch_loss)
        fold_train_loss.append(epoch_loss)

        #Validation
        with torch.no_grad():
            model.eval()
            valid_loss = []
            valid_true = []
            valid_pred = []
            for step, (batch_x, batch_y, batch_time, batch_static, batch_weight, batch_pdid) in enumerate(test_loader):
                batch_x = batch_x.to(device)
                batch_y = batch_y.to(device)
                batch_time = batch_time.to(device)
                batch_static = batch_static.to(device)
                batch_weight = batch_weight.to(device)
                
                valid_output, _,_ = model(batch_x, batch_time) #B T 1
                loss = get_loss(valid_output, batch_y.unsqueeze(-1), batch_weight.unsqueeze(-1))
                valid_loss.append(loss.cpu().detach().numpy())

                for m, t, p in zip(batch_weight.unsqueeze(-1).cpu().numpy().flatten(), batch_y.cpu().numpy().flatten(), valid_output.cpu().detach().numpy().flatten()):
                    if np.equal(m, 1):
                        valid_true.append(t)
                        valid_pred.append(p)

            valid_loss = np.mean(valid_loss)
            fold_valid_loss.append(valid_loss)
            valid_pred = np.array(valid_pred)
            valid_pred = np.stack([1 - valid_pred, valid_pred], axis=1)
            ret = metrics.print_metrics_binary(valid_true, valid_pred, verbose=0)

            if each_epoch % 10 == 0:
                print('Fold %d, epoch %d: Loss = %.4f Valid loss = %.4f roc = %.4f'%(fold_count, each_epoch, fold_train_loss[-1], fold_valid_loss[-1], ret['auroc']))

            cur_auroc = ret['auroc']
            if cur_auroc > best_auroc:
                best_auroc = cur_auroc
                best_auprc = ret['auprc']
                best_minpse = ret['minpse']
                state = {
                    'net': model.state_dict(),
                    'optimizer': optimizer.state_dict(),
                    'epoch': each_epoch
                }
                torch.save(state, file_name+'_'+str(fold_count))
                
                if cur_auroc > global_best:
                    global_best = cur_auroc
                    state = {
                        'net': model.state_dict(),
                        'optimizer': optimizer.state_dict(),
                        'epoch': each_epoch
                    }
                    torch.save(state, file_name)
                    print('------------ Save best model - AUROC: %.4f ------------'%cur_auroc)
                    
    auroc.append(best_auroc)
    auprc.append(best_auprc)
    minpse.append(best_minpse)
    total_train_loss.append(fold_train_loss)
    total_valid_loss.append(fold_valid_loss)

print('auroc %.4f(%.4f)'%(np.mean(auroc), np.std(auroc)))
print('auprc %.4f(%.4f)'%(np.mean(auprc), np.std(auprc)))
print('minpse %.4f(%.4f)'%(np.mean(minpse), np.std(minpse)))    
```
